```python
test_acc = []
total_nodes = 10
def parse_logs(numRuns, input_file_directory):
    for i in range(0, numRuns):

        fname = input_file_directory + "/log_0_" + str(total_nodes) + ".log"
        lines = [line.rstrip('\n') for line in open(fname)]
        for line in lines:
            idx = line.find("Test Acc")
            if idx != -1:
                timestamp = line[7:20]
                test_acc.append(line[(idx + 12):(idx+19)])

    return test_acc

        

parse_logs(1 , "../../Fed_Sys/Fed_log" )

for value in test_acc:
    test_acc[test_acc.index(value)] = float(value)
    
print("test_acc after convert \n" , test_acc)
```

```python
def Plot(epsilon2,epsilon4, without_epsilon ) :
  # plot train loss
  fig, ax = plt.subplots()
  x_axis = np.arange(1, len(epsilon2)+1)
  y_axis_loss1 = np.array(epsilon2)
  y_axis_loss2 = np.array(epsilon4)
  y_axis_loss3 = np.array(without_epsilon)
  ax.plot(x_axis, y_axis_loss1 , label = 'epsilon2')
  ax.plot(x_axis, y_axis_loss2 , label = 'epsilon4')
  ax.plot(x_axis, y_axis_loss3 , label = 'without_epsilon')
  plt.legend()

  ax.set(xlabel='Communication Rounds', ylabel='Train Loss',
       title="CNN net on IID-Mnist dataset")
  ax.grid()
#   fig.savefig("CNN net on IID-Mnist dataset"+'.jpg', format='jpg')

Plot(epsilon2,epsilon4, without_epsilon ) 
```
